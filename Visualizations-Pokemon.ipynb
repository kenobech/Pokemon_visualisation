#### Univariate Visualisation

```python
# create a count plot using generation_id
plt.figure(figsize=(10, 8))
sns.countplot(data=df, x='generation_id')
plt.title("Count of Pokemon by Generation")
plt.xlabel('Counts')
plt.ylabel('Generation Id')
plt.show();
```

From the above visualisations, species with generation_id 5 has got the highest counts while species with generation_id 6 has got the lowest count.

```python
# create a histogram
plt.figure(figsize=(10,8))
sns.histplot(df.attack)
plt.title('Attack Histogram')
plt.xlabel('Count')
plt.ylabel('Attack')
plt.show();
```

The attack histogram shows that most of the attacks occur at 75, even though there is an outlier in 50. The least attacks are at 175.

```python
# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['weight'])
plt.title('Weight Distribution')
plt.xlabel('Weight')
plt.show();
```

The weight distribution is highly concentrated between weights of 200-400.

```python
# Histogram for Height Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['height'], kde=True)
plt.title('Height Distribution')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show();
```

### Bi-variate Visualisation

```python
# Scatter plot between 'height' and 'weight' with regression line
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='height', y='weight', hue='species')
plt.title('Height vs. Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend(title='Species')
plt.show();
```

```python
# Heatmap to visualize the correlation between numeric columns
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
plt.yticks(rotation=45)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show();
```

```python
# Pair plot for selected numeric columns
selected_columns = ['height', 'weight', 'attack', 'defense']
sns.pairplot(df[selected_columns], diag_kind='kde')
plt.suptitle('Pair Plot of Selected Numeric Columns', y=1.02)
plt.show();
```

### Multi-variate visualisations

```python
# Create a pair plot
selected_columns = ['height', 'weight', 'attack', 'defense', 'speed', 'special-attack', 'special-defense']
sns.pairplot(df[selected_columns], diag_kind='kde')
plt.suptitle('Pair Plot of Selected Columns', y=1.02)
plt.show();
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
x = df['attack']
y = df['defense']
z = df['speed']
c = df['weight']
scatter = ax.scatter(x, y, z, c=c, cmap='viridis', marker='o')
plt.colorbar(scatter, label='Weight')

ax.set_xlabel('Attack')
ax.set_ylabel('Defense')
ax.set_zlabel('Speed')

plt.title('3D Scatter Plot with Color Mapping')
plt.show();
```
